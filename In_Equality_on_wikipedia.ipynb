# **what's on data...!**

* **LANGUAGE** --> List of languages on which articles are written
* **ARTICLE**--> Total number of article written in that language
* **TOTAL PAGE** --> Total number of pages in both article and non-article
* **EDITS** --> Number of edit's made in that article's
* **ADMINS** --> Access restricted Wikipedia software functions: deleting articles and files, protecting and unprotecting pages, and blocking and unblocking users
* **USERS** --> Number of user accounts, regardless of current activity not the amount of people or devices using Wikipedia.
* **ACTIVE USERS** --> Who have made at least one edit in the last thirty days
* **FILES** --> Number of locally uploaded files
* **DEPTH** --> Rough indicator of a Wikipedia's quality, showing how frequently its articles are updated. It does not refer to academic quality
* **PERCENTAGE** --> Percentage which each language contributes to articles in wikipedia

# **Computing lorenz curve to calculate the gini coefficient of the languages**

```python
# Calculate the total number of contributions
total_contributions = df['Articles'].sum()

# Compute the proportion of contributions
df['Contribution Proportion'] = df['Articles'] / total_contributions

# Sort the language editions in descending order of contribution proportion
df = df.sort_values('Contribution Proportion', ascending=False)

# Calculate the cumulative proportion of contributions
df['Cumulative Proportion'] = df['Contribution Proportion'].cumsum()

# Calculate the sum of the current cumulative proportions
current_sum = df['Cumulative Proportion'].sum()

# Normalize the cumulative proportions
normalized_proportions = df['Cumulative Proportion'] / current_sum

# Multiply the normalized proportions by 1 to achieve the correct sum
corrected_proportions = normalized_proportions * 1

# Update the 'Cumulative Proportions' column in the DataFrame with the corrected values
df['Cumulative Proportion'] = corrected_proportions

# Calculate the Lorenz curve
df['Lorenz Curve'] = df['Cumulative Proportion'] / df['Cumulative Proportion'].max()

# Plot the Lorenz curve
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Perfect Equality')
plt.plot(df['Cumulative Proportion'], df['Lorenz Curve'], label='Lorenz Curve')

# Set plot title and labels
plt.title('Lorenz Curve')
plt.xlabel('Cumulative Proportion of Contributions')
plt.ylabel('Cumulative Proportion of Users')

# Add legend
plt.legend()

# Display the plot
plt.show()

# Calculate the area between the Lorenz curve and the line of perfect equality
area_between_curve_and_line = df['Lorenz Curve'].diff().fillna(0).mul(df['Cumulative Proportion'])

# Calculate the Gini coefficient
gini_coefficient = 1 - 2 * area_between_curve_and_line.sum()

print('Gini Coefficient:', gini_coefficient)
```

```python
import matplotlib.pyplot as plt

gini_coeff = []

# Iterate over each language edition
for index, row in df.iterrows():
    edition_name = row['Language']

    # Calculate the total number of contributions for the edition
    total_contributions = row['Articles']

    # Compute the proportion of contributions
    df.loc[index, 'Contribution Proportion'] = row['Articles'] / total_contributions

    # Sort the language editions in descending order of contribution proportion
    edition = df.sort_values('Contribution Proportion', ascending=False)

    # Calculate the cumulative proportion of contributions
    edition['Cumulative Proportion'] = edition['Contribution Proportion'].cumsum()

    # Calculate the sum of the current cumulative proportions
    current_sum = edition['Cumulative Proportion'].sum()

    # Normalize the cumulative proportions
    normalized_proportions = edition['Cumulative Proportion'] / current_sum

    # Multiply the normalized proportions by 1 to achieve the correct sum
    corrected_proportions = normalized_proportions * 1

    # Update the 'Cumulative Proportions' column in the DataFrame with the corrected values
    edition['Cumulative Proportion'] = corrected_proportions

    # Calculate the Lorenz curve
    edition['Lorenz Curve'] = edition['Cumulative Proportion'] / edition['Cumulative Proportion'].max()

    # Plot the Lorenz curve
    plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Perfect Equality')
    plt.plot(edition['Cumulative Proportion'], edition['Lorenz Curve'], label='Lorenz Curve')

    # Set plot title and labels
    plt.title(f'Lorenz Curve - {edition_name}')
    plt.xlabel('Cumulative Proportion of Contributions')
    plt.ylabel('Cumulative Proportion of Users')

    # Add legend
    plt.legend()

    # Display the plot
    plt.show()

    # Calculate the area between the Lorenz curve and the line of perfect equality
    area_between_curve_and_line = edition['Lorenz Curve'].diff().fillna(0).mul(edition['Cumulative Proportion'])

    # Calculate the Gini coefficient
    gini_coefficient = 1 - 2 * area_between_curve_and_line.sum()


    print(f'Gini Coefficient for {edition_name}: {gini_coefficient}')

    #append(gini_coefficient, gini_coefficients)

    gini_coeff.append({'Language': edition_name, 'Gini Coefficient': gini_coefficient})

```

```python
df_gini = pd.DataFrame(gini_coeff)
df_gini
```

```python
plt.plot(df_gini['Language'], df_gini['Gini Coefficient'], label='G', marker='o')
plt.title('Gini Coefficients by Language')
plt.xlabel('Language')
plt.ylabel('Gini Coefficient')
plt.xticks(rotation=90)
plt.show()
```
